# Exercise 3 (Timing Stuff)

```python
squares = [i**2 for i in range(10)]
print(squares)
```

```python
%%timeit
squares = [i**2 for i in range(10)]
```

```python
squares = list()
for i in range(10):
    squares.append(i**2)
print(squares)
```

```python
%%timeit
squares = list()
for i in range(10):
    squares.append(i**2)
```

1.01 μs and 1.29 μs are the respective means of each run. 77 ns and 110 ns are the respective standard deviations for each run. 7 runs means that the code was ran 7 times, and 1,000,000 loops each means that for each run, the code was looped 1,000,000 times.
