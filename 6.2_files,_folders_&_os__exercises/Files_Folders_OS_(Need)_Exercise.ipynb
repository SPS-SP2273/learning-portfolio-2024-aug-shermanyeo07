## Task 4

```python
def rename_my_file(old_file_name):
    for key in month_num:
        if key in old_file_name:
            new_file_name = old_file_name.replace(key, month_num[key]).replace('_', '-')
            os.rename(old_file_name, new_file_name)
            print(f'{old_file_name} ----> {new_file_name}')
```

```python
files = glob.glob('os-collaboration-exercise-data/os-collaboration-exercise-data/*.txt')
```

```python
for old_file_name in files:
    try:
        rename_my_file(old_file_name)
        print(f'Renaming {old_file_name}')
    except FileExistsError:
        print(f'Renaming already completed; skipping rename')
```

## Task 7

```python
cities_lowercase = ['cairo', 'dhaka', 'jakarta', 'karachi', 'manila',
 'melbourne', 'osaka', 'shanghai', 'singapore', 'tokyo']

loc = 'os-collaboration-exercise-data'

for c in cities_lowercase:
    files = glob.glob(os.path.join(loc, loc, f'*{c}.txt'))
    for txt in files:
        path_to_move = os.path.join(f'{c}', os.path.basename(txt))
        if os.path.exists(path_to_move):
            print(f'{path_to_move} already exists, did not move') # was testing to see if code can recognise if file already exists
        else:
            shutil.move(txt, path_to_move)
```
